```python
import pandas as pd
import numpy as np
from pylab import rcParams
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt

import sklearn
from sklearn.cluster import DBSCAN
import sklearn.datasets as data
from collections import Counter
from pandas import DataFrame
%matplotlib inline
```

```python
 #blobs, _ = data.make_blobs(n_samples=100, centers=[(-0.75,2.25), (1.0, 2.0)], cluster_std=0.01)
X,_= data.make_moons(n_samples=100, noise=0.1) 
```

```python
df = DataFrame(dict(x=X[:,0],y=X[:,1]))
fig,ax= plt.subplots()
df.plot(ax=ax, kind='scatter', x='x', y='y')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=.5, min_samples=15).fit(X)
cluster= clustering.labels_
```

```python
len(set(cluster))
```

```python
def show_clusters(X,cluster):
    df = DataFrame(dict(x=X[:,0], y=X[:,1], label=cluster))
    colors = {-1:'red', 0:'blue', 1:'orange', 2:'green', 3:'skyblue', 4:'black'}
    fig, ax = plt.subplots(figsize=(10,10))
    grouped = df.groupby('label')
    for key, group in grouped:
        group.plot(ax=ax, kind='scatter',x='x',y='y', label=key, color=colors[key])
    plt.show()
```

```python
show_clusters(X, cluster)
```
